```python
sty_to_tui = dict((el, srdef[srdef.STYorRL == el][["UI"]].values[0][0]) for el in sty)
```

```python
sty_to_tui
```

```python
with open('STY_to_TUI.pickle', 'wb') as handle:
    pickle.dump(sty_to_tui, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
sty_to_cuis = dict((el, []) for el in sty)
for el in sty_to_cuis:
    tui = sty_to_tui[el]
    sty_to_cuis[el] = mrsty[mrsty.TUI == tui].CUI.tolist()
```

```python
with open('STY_to_CUIs.pickle', 'wb') as handle:
    pickle.dump(sty_to_cuis, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
cui_to_stys = dict((el, [])for el in mrsty.CUI.drop_duplicates())
for sty in sty_to_cuis:
    for cui in sty_to_cuis[sty]:
        cui_to_stys[cui] += [sty]
```

```python
cui_to_stys
```

```python
with open('CUI_to_STYs.pickle', 'wb') as handle:
    pickle.dump(cui_to_stys, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
sty_to_concepts = sty_to_cuis
```
